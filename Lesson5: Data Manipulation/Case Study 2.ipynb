```python
import pandas as pd
import numpy as np
df=pd.read_csv('Salaries.csv')
df
```

```python
data=df.loc[:,['Year','TotalPay']]
print(data.groupby('Year').mean())
```

```python
data=df.loc[:,['JobTitle','Year','TotalPay']]
print (data.max())
```

```python
data=df.loc[:,['Year','OvertimePay']]
print (data.groupby('Year').sum())
```

```python
data=df.loc[:,['Year','JobTitle','TotalPayBenefits']]
s_data=data[data['Year']==2014]
print(s_data.groupby('JobTitle')['TotalPayBenefits'].nlargest(5))
```

```python
data=df.loc[:,['EmployeeName','Year','TotalPayBenefits']]
print (data.nlargest(5,'TotalPayBenefits'))
```
