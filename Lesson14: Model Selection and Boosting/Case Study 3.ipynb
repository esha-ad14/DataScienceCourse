```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Question 1
df = pd.read_csv("letterCG.bin", sep=' ')
df
```

```python
df.drop(["Unnamed: 5", "Unnamed: 18", "yegvx"], axis=1, inplace=True)
df.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,1:]
y = df["Class"]
```

```python
x
```

```python
y
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, random_state = 10, test_size = 0.20)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
#Question 2
dt= DecisionTreeClassifier(max_depth=1)
ada = AdaBoostClassifier(base_estimator=dt,n_estimators=400,learning_rate=1,algorithm='SAMME')

ada.fit(train_x,train_y)
prediction = ada.predict(test_x)
metrics.accuracy_score(prediction, test_y)
```

```python
#Question 3
dt= DecisionTreeClassifier(max_depth=2)
ada = AdaBoostClassifier(base_estimator=dt,n_estimators=400,learning_rate=1,algorithm='SAMME')

ada.fit(train_x,train_y)
prediction = ada.predict(test_x)
metrics.accuracy_score(prediction, test_y)
```
