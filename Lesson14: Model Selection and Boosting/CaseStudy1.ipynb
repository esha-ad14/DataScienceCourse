```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bio-degradabale-data.csv",sep=';')
df
```

```python
df.columns=range(42)
df
```

```python
x = df.iloc[:,0:41]
y = df[41]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, random_state = 10, test_size = .20)
```

```python
logR = LogisticRegression()
logR.fit(train_x, train_y)
prediction = logR.predict(test_x)
print(metrics.accuracy_score(prediction, test_y))
```

```python
print(cross_val_score(logR, x, y, cv=10, scoring='accuracy').mean())
```
