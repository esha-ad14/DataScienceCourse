```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Project_Data_1.csv", index_col=0, decimal=",")
df
```

```python
df.isna()
```

```python
df.dropna()
```

```python
from sklearn.decomposition import PCA
data = df.iloc[:,0:]
data
```

```python
pca = PCA(n_components=2)
dataPCA= pca.fit(data).transform(data)
```

```python
print("Variance Ratio:",pca.explained_variance_ratio_)
```

```python
ndata = pd.DataFrame(dataPCA, columns=["pca_1", "pca_2"])
ndata.index = df.index
ndata.head()
```

```python
from sklearn.cluster import KMeans
wcss = [] 
for i in range(1,11):
    km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=10)
    km.fit(ndata)
    wcss.append(km.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title("Elbow Method")
plt.xlabel("Clusters")
plt.ylabel("wcss")
plt.show()
```

```python
km = KMeans(n_clusters=3)
```

```python
km.fit(ndata)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
ndata["cluster"] = km.labels_
ndata
```

```python
import seaborn as sns
sns.lmplot('pca_1', 'pca_2', data=ndata, hue='cluster', palette='coolwarm', size=5, aspect=1, fit_reg=False)
```

```python
df["cluster"] = km.labels_
```

```python
df.sort_values(["cluster"])
```

```python
ndata.sort_values(["pca_1", "pca_2"])
ndata.to_csv("output.csv", index=True)
```
