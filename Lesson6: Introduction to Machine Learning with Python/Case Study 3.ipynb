```python
df.columns
x = df[['Avg_Session_Length', 'Time_on_App', 'Time_on_Website', 'Length_of_Membership']]
y = df['Yearly_Amount_Spent']

```

```python
#Question 5
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=85)
X_train, X_test, y_train, y_test
```

```python
#Question 6
from sklearn.linear_model import LinearRegression 
linm = LinearRegression()
linm.fit(X_train,y_train)
prediction = linm.predict(X_test)
plt.scatter(y_test,prediction, color="lightblue")
plt.xlabel('Y Test values')
plt.ylabel('Predicted values')

#some of the values match as they are both correlated 
```

```python
#Question 7
from sklearn import metrics
print ('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```
