```python
#Data Loading
#a)
import numpy as np
import pandas as pd
df=pd.read_csv("prisoners.csv")
print(df.head())
print(df.tail())
```

```python
#b)
df.describe()
#those with 0 inmates are probably repeated rows or don't have recorded data
```

```python
#Data Manipulation
#a)
#df["total_benefitted"] = df["No. of Inmates benefitted by Elementary Education"] + df["No. of Inmates benefitted by Adult Education"] + df["No. of Inmates benefitted by Higher Education"] + df["No. of Inmates benefitted by Computer Course"]
#df
df["total_benefited"]=df.sum(axis=1)
df

```

```python
#b)
df = df.append(df.sum(), ignore_index=True)
df.at[35, "STATE/UT"] = "totals"

df.at[35, "YEAR"] = 2000

df.head()
```

```python
#plotting
#a)
import matplotlib.pyplot as plt
import numpy as np
xlabels = df['STATE/UT'].values
plt.figure(figsize=(20, 10))
plt.xticks(np.arange(xlabels.shape[0]), xlabels, rotation = 'vertical', fontsize = 18)
plt.xticks
plt.bar(np.arange(df.values.shape[0]),df['total_benefited'],align = 'edge')
```

```python
#b)
labels = df["STATE/UT"]
values = df["total_benefited"]
plt.figure(figsize=(25,25))
plt.pie(values, labels = labels, autopct='%1.1f%%', shadow=True)
plt.title("STATE/UT vs Total_inmates_benefitted")
plt.show()
```
