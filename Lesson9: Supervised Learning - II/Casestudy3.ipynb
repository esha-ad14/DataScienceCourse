```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, ShuffleSplit, GridSearchCV
from sklearn import svm
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
#Question 1
df = pd.read_csv("College.csv")
df

labelencoder = LabelEncoder()
df["Private"] = labelencoder.fit_transform(df["Private"])
df
```

```python
#Question 2
x = df.iloc[:, 1:]
y = df["Private"]

train_x, test_x, train_y, test_y = train_test_split(
    x, y, random_state=10,test_size=0.25)
```

```python
#Question 3
svm = svm.LinearSVC()
svm.fit(train_x, train_y)
predictedVals = svm.predict(test_x)

print("Accuracy Score:",metrics.accuracy_score(predictedVals, test_y))
```

```python
#Question 4
from sklearn.svm import LinearSVC, SVC
scaler = StandardScaler().fit_transform(x)
scaler = pd.DataFrame(x, columns=x.columns)

x = scaler
y = df["Private"]

train_x, test_x, train_y, test_y = train_test_split(
    x, y, random_state=10, test_size=0.25)

svm = LinearSVC()
svm.fit(train_x, train_y)

predictedVals = svm.predict(test_x)
metrics.accuracy_score(predictedVals, test_y)
```

```python
#Question 5
parameter = [
    {'C': [1, 10, 100, 1000], 'kernel': ['poly']},
    {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
    {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['rbf']},
]

# Create a classifier object with the classifier and parameter candidates
cv = ShuffleSplit()
clf = GridSearchCV(estimator=SVC(max_iter=1000),
                   param_grid=parameter, n_jobs=-1, cv=cv)

# Train the classifier on data1's feature and target data
clf.fit(train_x, train_y)

# View the accuracy score
print('Best score for data1:', clf.best_score_)

# View the best parameters for the model found using grid search
print('Best C:', clf.best_estimator_.C)
print('Best Kernel:', clf.best_estimator_.kernel)
print('Best Gamma:', clf.best_estimator_.gamma)
```
