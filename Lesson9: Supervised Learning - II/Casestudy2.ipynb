```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
#Question 1
df = pd.read_csv("run_or_walk.csv")
df
```

```python
#Question 2
x = df.iloc[:, 5:]
y = df["activity"]
```

```python
#Question 3
train_x, test_x, train_y, test_y = train_test_split(
    x, y, random_state=10, test_size=0.25)

gaussian = GaussianNB()
gaussian.fit(train_x, train_y)

predictedVals = gaussian.predict(test_x)

print("Accuracy Score:",metrics.accuracy_score(predictedVals, test_y))

print("Classification Score")
print(metrics.classification_report(predictedVals, test_y))

```

```python
#Quedtion 4
x1 = df.iloc[:, 5:8]
y1 = df["activity"]

train_x1, test_x1, train_y1, test_y1 = train_test_split(
    x1 , y1, random_state=10, test_size=0.25)

gaussian.fit(train_x1, train_y1)
predictedVals1 = gaussian.predict(test_x1)

print("Accuracy Score:",metrics.accuracy_score(predictedVals1, test_y1))

print("Classification Score")
print(metrics.classification_report(predictedVals1, test_y1))


# Gyro as Independent variables
x2 = df.iloc[:, 8:]
y2 = df["activity"]

train_x2, test_x2, train_y2, test_y2 = train_test_split(
    x2, y2, random_state=10, test_size=0.25)

gaussian.fit(train_x2, train_y2)
predictedVals2 = gaussian.predict(test_x2)

print("Accuracy Score:",metrics.accuracy_score(predictedVals2, test_y2))

print("Classification Score")
print(metrics.classification_report(predictedVals2, test_y2))
```
