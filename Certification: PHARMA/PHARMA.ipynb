```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("train.csv")
train
```

```python
test = pd.read_csv("test.csv")
test
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train["species"] = le.fit_transform(train["species"])
train
```

```python
x = train.iloc[:,1:]
y = train["species"]
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, random_state = 10, test_size = 20)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(train_x, train_y)

prediction_rf = rf.predict(test_x)
print("Radom Forest Accuracy Score : ", metrics.accuracy_score(prediction_rf, test_y))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(train_x, train_y)

prediction_dt = dt.predict(test_x)
print("Decision Tree Accuracy Score : ", metrics.accuracy_score(prediction_dt, test_y))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(train_x, train_y)

prediction_gnb = gnb.predict(test_x)
print("Naive Bayes Accuracy Score : ", metrics.accuracy_score(prediction_gnb, test_y))
```

```python
from sklearn.svm import LinearSVC

svc = LinearSVC()
svc.fit(train_x, train_y)

prediction_svc = svc.predict(test_x)
print("SVM Accuracy Score : ", metrics.accuracy_score(prediction_svc, test_y))
```

```python
prediction_test = gnb.predict(test)

test["species"] = le.inverse_transform(prediction_test)
test
```

```python
test.to_csv("Output.csv", index=True)
```
